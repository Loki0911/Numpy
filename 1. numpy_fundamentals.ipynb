## 6. Indexing/slicing

```python
a1 = np.arange(10,dtype = np.int32)
a2 = np.arange(12, dtype = float).reshape(3,4)
a3 = np.arange(8).reshape(2,2,2)

print(a1,a2,a3, sep ='\n\n')
```

```python
print(a2)
```

```python
#accessing 6 form a2
a2[1][2]    # -> a2[row][column]  for 2D
```

```python
print(a3)
```

```python
#to access 5 form a3 
a3[1][0][1]  # -> a3[pos of 2D array in which item exists][row][colum]
```

```python
#slicing

print(a1)
```

```python
a1[2:-1:2]  # -> same as python list
```

```python
a2
```

```python
# 5 6 9 10
a2[1:,1:3]
```

```python
# 0 3 8 11
a2[::2,::3]
```

```python
# 1 3 9 11

a2[::2,1::2]
```

```python
# 4 7
a2[1,::3]
```

```python
# 1 2 3 5 6 7
a2[0:2,1:]
```

```python
# 3D case
a3 = np.arange(27).reshape(3,3,3)
a3
```

```python
a3[1]
```

```python
a3[::2]
```

```python
# 3 4 5
a3[0,1,:]
```

```python
# 10 13 16
a3[1,:,1]
```

```python
# 22 23 25 26
a3[2,1:,1:]
```

```python
# 0 2 18 20

a3[::2,0,::2]
```

## 9. Stacking

```python
# horizontal stacking -> add columns
a4 = np.arange(12).reshape(3,4)
a5 = np.arange(12,24).reshape(3,4)

np.hstack((a4,a5))
```

```python
# vertical Stacking -> add rows
np.vstack((a4,a5))
```

## 10. Splitting

```python
# horizontal spliting
arr = np.hstack((a4,a5))
arr
```

```python
# horizontal splitting -> splits columns
np.hsplit(arr,2)   # -> a4 in 2 parts
```

```python
a4
```

```python
# vertical splitting  -> split rows
np.vsplit(a4,3)
```

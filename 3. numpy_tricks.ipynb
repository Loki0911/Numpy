## `6. np.where` 

The numpy.where() function returns the indeces of elements in an input array where the given condition is satisfied

```python
a
```

```python
# findall indices with the value greater than 50

np.where(a>50)
```

```python
#replace all values > 50 with 0
#np.where(condition,True,False)

np.where(a>50,0,a) 
```

```python
# to replace evens with 0

np.where(a%2 == 0,0,a)
```

## `11. ` np.corrcoef

return Pearson Product-moment correlation coefficients

```python
salary = np.array([20000,40000,25000,35000,60000])
experience = np.array([1,3,2,4,2])

np.corrcoef(salary, experience)   

##              salary     experince
##  salary          1           0.25
##  experience      0.25           1

```
